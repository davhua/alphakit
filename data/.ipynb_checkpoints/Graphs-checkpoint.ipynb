```python
# libraries
import pandas as pd
import numpy as np
import urllib
from matplotlib import pyplot as pl
from datetime import date
from dateutil.relativedelta import relativedelta
%matplotlib inline
```

```python
# load data into pandas dataframes
with open('QUANDL-CHRIS-CME_S1.csv') as f:
	soyfutures = pd.read_csv(f, index_col='Date', parse_dates=True)

soyfutures['daily_rtn'] = pd.Series((soyfutures['Settle'] - soyfutures['Settle'].shift(-1)) / soyfutures['Settle'].shift(-1), index=soyfutures.
index)
soyfutures['cml_daily_rtn'] = pd.Series(soyfutures['daily_rtn'] + 1, index=soyfutures.index).cumprod()
soyfutures[['cml_daily_rtn']].plot()

# Soybean Futures Cumulative Daily Return Plot
```

```python
# load data into pandas dataframes
with open('QUANDL-CHRIS-CME_B01.csv') as f:
	soyoilfutures = pd.read_csv(f, index_col='Date', parse_dates=True)

soyoilfutures['daily_rtn'] = pd.Series((soyoilfutures['Settle'] - soyoilfutures['Settle'].shift(-1)) / soyoilfutures['Settle'].shift(-1), index=soyoilfutures.
index)
soyoilfutures['cml_daily_rtn'] = pd.Series(soyoilfutures['daily_rtn'] + 1, index=soyoilfutures.index).cumprod()
soyoilfutures[['cml_daily_rtn']].plot()

# Soybean Oil Futures Cumulative Daily Return Plot
```

```python
# load data into pandas dataframe
with open('QUANDL-CFTC-S_F_ALL.csv') as f:
	soycts = pd.read_csv(f, index_col='Date', parse_dates=True)

soycts['long_short_ratio'] = pd.Series.div(soycts['Total Reportable Longs'], soycts['Total Reportable Shorts'])
soycts[['long_short_ratio']].plot()

#Soybean Total Reportable Longs/Shorts Ratio
```
